### 4.2 What percentage of people report that they have an "average", "fit", or "athletic" body type?  
(sum all theree types)

```python
avg,fit,athletic =df[df["body_type"]=="average"],df[df["body_type"]=="fit"],df[df["body_type"]=="athletic"]
avg_prect,fit_prect,athletic_prect = 100*(len(avg)/len(df)),100*(len(fit)/len(df)),100*(len(athletic)/len(df))
total_precnt = avg_prect+fit_prect+athletic_prect
print(f"The percentage of people report that they have an average, fit, or athletic body type is {total_precnt}")
```

### (Bonus) How do you interpret the results of the previous question?

### 5.3 How many men shorter than 170cm are there in the data? How many women?

```python
shorts = df[df['height_(cm)']<170]
shorts_men = len(shorts[shorts["sex"]=="m"])
shorts_women = len(shorts[shorts["sex"]=="f"])
print(f"The amount of men shorter than 170cm is {shorts_men}. The amount of women shorter than 170cm is {shorts_women}")
```

```python
# If we want to check that we dont have any mistake lets compare between the total to the sum of woman +sum of men
sum_shorts = len(shorts)
print(shorts_men+shorts_women)
print(sum_shorts)
print("Ok , we are good :)")
```

### How do you interpret the results of the previous question?  
(regarding people under 170cm)

```python
women_precent = int(100*(shorts_women/sum_shorts))
print(f"Of all the Shorts people, {women_precent} percent are women" )
```
