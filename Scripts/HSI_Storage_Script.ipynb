### In this script i store my hyperspectral data. 

1. ```Store_HSI```  Function:

  After selecting the pixels of each leaf in the *HSview* software and exporting them to a CSV file, I upload the CSV files here, arrange the columns and insert column of label by leaf number. This way I can make sure that all the data I collected will be saved in the same format, which will make it easier to insert it into the model later. Then i save the arranged data.

2. ```Add_To_CSV``` Function:
  
  In this function I will:
  
  1. Group the HS data by mean leaf number.
  2. combine the samples of a specific date with the data that came from the file of gas exchange samples received from the LI6800 system.
  2. I will then take a number of relevant columns (parameters) such as light intensity, An, Tr etc. and add them together with the hyperspectral data to a CSV file that contains HS data + sorted LI-6800 data  and ready to perform model. This combined data contain only specific parameters of LI6800 (the parameters that relevant for predict An and Tr).

````python
import pandas as pd
import numpy as np
````

````python
def Store_HSI(path_to_file_right,
              path_to_file_left,
              array_right,
              array_left,
              path_to_save
              ):
  ############################ Right Side ############################
  df= df= pd.read_csv(path_to_file_right) # Read the file
  bands_array = np.arange(1,205,1)
  df['band'] = bands_array # band number
  df_t = df.T # Transpose 
  index = np.arange(0,len(df_t),1)
  df_t = df_t.set_index(index) # set to index
  df_t.rename(index={0:'wavelength'},inplace=True) # Replace first value in column
  all_data = df_t.iloc[1:-1] #all pixels values
  df_t.rename(index={1:'band'},inplace=True) # Replace second value in column
  # arange the data in the right order:
  df_t = pd.concat([df_t.iloc[:2],all_data])
  # # locate the bands in the right row:
  df_t.loc["band"] = np.arange(1,205,1) # band num
  
  # Rename the value in column by the array values
  array = array_right
  df_t['index'] = 0
  array1 = array.tolist()
  # replace 2 first values:
  df_t.loc["wavelength",'index'] = "wavelength"
  df_t.loc["band",'index'] = "band"
  # Fill the values of column by array:
  df_t.loc[array1[0]:array1[1],'index'] = 1
  df_t.loc[array1[1]+1:array1[2],'index'] = 2
  df_t.loc[array1[2]+1:array1[3],'index'] = 3
  df_t.loc[array1[3]+1:array1[4],'index'] = 4
  df_t.loc[array1[4]+1:array1[5],'index'] = 5
  df_t.loc[array1[5]+1:array1[6],'index'] = 6
  df_t.loc[array1[6]+1:array1[7],'index'] = 7
  df_t.loc[array1[7]+1:array1[8],'index'] = 8
  df_t.loc[array1[8]+1:,'index'] = 9
  df_t = df_t.set_index("index")
  df_t.index.name = None #remove the name of index col
  df_right = df_t #the right side of the Green Wall

  ############################ Left Side ############################
  df= pd.read_csv(path_to_file_left) # Read the file
  df['band'] = bands_array # band number
  df_t = df.T # Transpose 
  index = np.arange(0,len(df_t),1)
  df_t = df_t.set_index(index) # set to index
  df_t.rename(index={0:'wavelength'},inplace=True) # Replace first value in column
  all_data = df_t.iloc[1:-1] #all pixels values
  df_t.rename(index={1:'band'},inplace=True) # Replace second value in column
  # arange the data in the right order:
  df_t = pd.concat([df_t.iloc[:2],all_data])
  # # locate the bands in the right row:
  df_t.loc["band"] = np.arange(1,205,1) # band num
  
  # Rename the value in column by the array values
  array = array_left
  df_t['index'] = 0
  array1 = array.tolist()
  # replace 2 first values:
  df_t.loc["wavelength",'index'] = "wavelength"
  df_t.loc["band",'index'] = "band"
  # Fill the values of column by array:
  df_t.loc[array1[0]:array1[1],'index'] = 10
  df_t.loc[array1[1]+1:array1[2],'index'] = 11
  df_t.loc[array1[2]+1:array1[3],'index'] = 12
  df_t.loc[array1[3]+1:array1[4],'index'] = 13
  df_t.loc[array1[4]+1:array1[5],'index'] = 14
  df_t.loc[array1[5]+1:array1[6],'index'] = 15
  df_t.loc[array1[6]+1:array1[7],'index'] = 16
  df_t.loc[array1[7]+1:array1[8],'index'] = 17
  df_t.loc[array1[8]+1:,'index'] = 18
  df_t = df_t.loc[2:] #delete the 2 first rows (band and wavelengh)
  df_t = df_t.set_index("index")
  df_t.index.name = None
  df_left = df_t # the left side of the Green Wall

 ############################ Merge and Save ############################
  merged_df = pd.concat([df_right,df_left]) # one beneath the other
  merged_df_t = merged_df.T # transpose
  merged_df_t = merged_df_t.set_index('wavelength') #set index column
  merged_df_t.index.name = 'wavelength'
  merged_df = merged_df_t.T # transpose back with index
  # # Save the 2 into one excel with 2 sheets:
  with pd.ExcelWriter(path_to_save) as writer:  
    merged_df_t.to_excel(writer, sheet_name='Sheet_0')
    merged_df.to_excel(writer, sheet_name='Sheet_1')
  
  return merged_df
````

````python
def Add_To_CSV(df_path):
  
  ##### Read the data #####
  excel_file = pd.ExcelFile(df_path) #read the whole file
  sheet = excel_file.sheet_names[1] #take the second sheet
  df_raw = pd.read_excel(df_path, sheet_name=sheet) # read the file with the second sheet
  df_raw= df_raw.loc[1:] # delete first row
  df_raw.rename(columns = {df_raw.columns[0]:'leaf_hs',}, inplace = True) # Rename column of leaves

  ##### Group by leaf mean #####
  df_avg = df_raw.groupby(['leaf_hs']).mean()
  df_avg
  
  ##### Add date column and arange columns order #####
  '''
  Here I will add a date column based on the file name.
  Notice that your file should be in a specific format and change the index of the date string.

  In addition, I will change the order of the column.
  It is not necessary it is just the way that my data arranged already.
  '''
  # reindex:
  df_avg = df_avg.reset_index()

  # Move first column to the end
  new_cols = [col for col in df_avg.columns if col != 'leaf_hs'] + ['leaf_hs']
  df_avg = df_avg[new_cols]
  # date column:
  date = df_path[52:62] # Date by string index
  df_avg['date_hs'] = date
  df_avg['date_hs'] = pd.to_datetime(df_avg['date_hs'],format="%d_%m_%Y")
  
  ##### Save as csv by date#####
  
  df_avg.to_csv("/content/drive/MyDrive/Greenwall/Hyperspectral_data/Files_to_store/" + date +".csv")

  return df_avg
````
